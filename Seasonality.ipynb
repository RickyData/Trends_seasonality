```python
file['Date'] = pd.to_datetime(file['Date'])
file.info()
```

```python
import numpy as np

file['Year'] = file.Date.dt.year
file['Month'] = file.Date.dt.month
file['Week_day'] = file.Date.dt.weekday
file['Week_dayName'] = file.Date.dt.weekday_name
file['Week_year'] = file.Date.dt.weekofyear
file['Day'] = file.Date.dt.day
file['Returns'] = (file.Close/file.Close.shift(1)) - 1
file['Direction'] = np.where(file.Returns >0,'Bull', 'Bear')



file.head()
```

```python
file_A = file[['Date','Year', 'Month', 'Day', 'Week_day', 'Week_year', 'Open', 'High', 'Low', 'Close', 'Volume', 'Returns', 'Direction']]
```

```python
file_A = file_A.loc[file_A.Year == 2003:2017, :]
#file_A = file_A.loc[file_A.Year < 2018, :]
file_A.tail()
```

```python
file_A.set_index(file_A['Date'],inplace=True)
file_A.head()
```

```python
present = file_A.loc[file_A['Week_year']== 39,]
bulls = present.loc[present['Direction'] == 'Bull',:]
bears = present.loc[present['Direction'] == 'Bear',:]
```

```python
op_1 = bulls.loc[bulls['Week_day'] == 'Monday', :]
op_1
```

```python
op_2 = bears.loc[bears['Week_day'] == 'Friday', :]
op_2
```

```python
bull_Confidence =len(op_1)/(len(op_1) + len(op_2))
bull_Confidence
```

```python
bear_Confidence = len(op_2)/(len(op_2) + len(op_1))
bear_Confidence
```

```python
raw_data = {'student_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze', 'Jacon', 'Ryaner', 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'], 
        'test_score': [76, 88, 84, 67, 53, 96, 64, 91, 77, 73, 52, np.NaN]}
df = pd.DataFrame(raw_data, columns = ['student_name', 'test_score'])
```

```python
# Create a list to store the data
grades = []

# For each row in the column,
for row in df['test_score']:
    # if more than a value,
    if row > 95:
        # Append a letter grade
        grades.append('A')
    # else, if more than a value,
    elif row > 90:
        # Append a letter grade
        grades.append('A-')
    # else, if more than a value,
    elif row > 85:
        # Append a letter grade
        grades.append('B')
    # else, if more than a value,
    elif row > 80:
        # Append a letter grade
        grades.append('B-')
    # else, if more than a value,
    elif row > 75:
        # Append a letter grade
        grades.append('C')
    # else, if more than a value,
    elif row > 70:
        # Append a letter grade
        grades.append('C-')
    # else, if more than a value,
    elif row > 65:
        # Append a letter grade
        grades.append('D')
    # else, if more than a value,
    elif row > 60:
        # Append a letter grade
        grades.append('D-')
    # otherwise,
    else:
        # Append a failing grade
        grades.append('Failed')
        
# Create a column from the list
df['grades'] = grades
```
